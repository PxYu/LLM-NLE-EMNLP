```python
import json
from src.AdaptiveBinning import AdaptiveBinning, AdaptiveBinningForRegression
```

```python
all_data = []
with open(f"processed_data/monot5-base-msmarco.jsonl", "r") as file:
    for line in file:
        item = json.loads(line)
        all_data.append(item)
```

```python
results = []
for x in all_data:
    results.append((x['score'], bool(x['label'])))
```

```python
AECE, AMCE, cof_min, cof_max, confidence, accuracy = AdaptiveBinning(results)
```

```python
for a, b, c, d in zip(cof_min, cof_max, confidence, accuracy):
    print(f"[{a:.3f},{b:.3f}] -- conf: {c:.3f} -- acc: {d:.3f}")
```
